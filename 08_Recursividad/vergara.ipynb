*1)  Crea una funciÃ³n recursiva que calcule el factorial de un nÃºmero. Luego, utiliza esa funciÃ³n para calcular y mostrar en pantalla el factorial de todos los nÃºmeros enteros entre 1 y el nÃºmero que indique el usuario*

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return factorial(n - 1) * n

factorial(5)
```

*2) Crea una funciÃ³n recursiva que calcule el valor de la serie de Fibonacci en la posiciÃ³n indicada. Posteriormente, muestra la serie completa hasta la posiciÃ³n que el usuario especifique.*

```python
def fibonacci(n):
    if n == 0 or n == 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```

*3) Crea una funciÃ³n recursiva que calcule la potencia de un nÃºmero base elevado a un exponente, utilizando la fÃ³rmula ð‘›ð‘š = ð‘› âˆ— ð‘›(ð‘šâˆ’1). Prueba esta funciÃ³n en un algoritmo general.*

```python
def potencia(n, m):
    if m == 0:
        return 1
    else:
        return n * potencia(n, m -1)

potencia(2, 3)
```

*4) Crear una funciÃ³n recursiva en Python que reciba un nÃºmero entero positivo en base
decimal y devuelva su representaciÃ³n en binario como una cadena de texto*

```python
def decimal_a_binario(n):
    if n == 0:
        return ""  
    else:
        return decimal_a_binario(n // 2) + str(n % 2)

print(decimal_a_binario(10))
```
